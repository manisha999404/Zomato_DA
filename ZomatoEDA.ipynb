```python
df['Average Cost for two']= ( df['Average Cost for two'] .astype(str) .str.replace(',', '') .astype(float))
```

```python
df['Has Table booking']=df['Has Table booking'].str.strip().str.title()
df['Has Online delivery']=df['Has Online delivery'].str.strip().str.title()
df['Is delivering now']=df['Is delivering now'].str.strip().str.title()
df['Switch to order menu']=df['Switch to order menu'].str.strip().str.title()
```

```python
print(df['Has Table booking'].value_counts())
print(df['Has Online delivery'].value_counts())
```

```python
df.drop(['Is delivering now', 'Switch to order menu', 'Rating color', ], axis=1, inplace=True)
```

```python
df
```

```python
import re

# Strip whitespace
df['Restaurant Name'] = df['Restaurant Name'].str.strip()

# Remove special characters
df['Restaurant Name'] = df['Restaurant Name'].apply(
    lambda x: re.sub(r'[^A-Za-z0-9\s]', '', str(x))
)

# Collapse multiple spaces into one
df['Restaurant Name'] = df['Restaurant Name'].str.replace(r'\s+', ' ', regex=True)
```

```python
df
```

```python

df['Cuisines'] = df['Cuisines'].fillna("Unknown")


df.drop_duplicates(inplace=True)
```

```python
df = df.assign(Cuisine=df['Cuisines'].str.split(',')).explode('Cuisine')
df['Cuisine'] = df['Cuisine'].str.strip()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Votes'], bins=30, kde=True)
plt.title("Votes Distribution")
plt.show()
```

```python
sns.countplot(x='Price range', data=df)
plt.title("Price Range Distribution")
plt.show()
```

```python
top_cuisines = df['Cuisine'].value_counts().head(10)
sns.barplot(x=top_cuisines.values, y=top_cuisines.index)
plt.title("Top 10 Cuisines")
plt.show()
```

```python
city_rating = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
city_rating.plot(kind='bar', figsize=(10,5), title="Top 10 Cities by Avg Rating")
plt.ylabel("Average Rating")
plt.show()

city_counts = df['City'].value_counts().head(10)
sns.barplot(x=city_counts.values, y=city_counts.index)
plt.title("Top 10 Cities by Number of Restaurants")
plt.show()
```
